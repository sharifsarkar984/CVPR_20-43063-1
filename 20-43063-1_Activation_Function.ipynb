IMPLEMENTATION OF DIFFERENT ACTIVATION FUNCTION

An activation function is a non-linear mathematical function used in artificial neural networks to introduce non-linearity into the output of a neural network node. Activation_Function

```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-9, 10, 1)
print(x)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

1.STEP FUNCTION

The step activation function is a simple type of activation function used in artificial neural networks, typically for binary classification problems. It maps an input value to a binary output value, either 0 or 1, based on whether the input value is greater or less than a predefined threshold value.A step function also includes the sum or product of two-step functions.

The step function has a discontinuous output, jumping suddenly from 0 to 1 as the input value crosses the threshold. This makes it less suitable for training neural networks using gradient-based optimization methods, as the derivative of the step function is 0 almost everywhere, making it difficult to update the weights of the network during backpropagation.

```python
# def step(x):
#     if x > 0.5:
#         return 1
#     else:
#         return 0

# # y = []
# # for i in x:
# #     y.append(step(i))
# y = list(map(step, x))

y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

2.SIGMOID FUNCTION

The sigmoid activation function is a mathematical function commonly used in artificial neural networks to introduce non-linearity to the output of a node. It is a smooth, S-shaped function that maps any input value to a value between 0 and 1.

The sigmoid function has a few important properties that make it useful for neural networks. First, it is a continuous function, which means that it has a derivative at every point. This makes it possible to use gradient-based optimization techniques to train the neural network. Second, it is a non-linear function, which means that it can introduce non-linearities into the output of the neural network, allowing it to learn complex patterns in the data.

```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

3.TANH FUNCTION

The hyperbolic tangent (tanh) activation function is a popular mathematical function used in artificial neural networks to add nonlinearity to a node's output. It is similar to the sigmoid function, but it maps input values to a range between -1 and 1, making it useful for problems with meaningful negative values.

The tanh function, like the sigmoid function, is a continuous and non-linear function that can be used in neural networks. It also has the property of being zero-centered, which means that its output is centered around zero, which can aid in network convergence during training.

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

4.RELU - RECTIFIED LINEAR UNITS

The rectified linear unit (ReLU) activation function is a common mathematical function used in artificial neural networks to add nonlinearity to a node's output. It is a straightforward function that returns the input value if it is positive and zero otherwise.

The ReLU function is superior to other activation functions in several ways. For starters, it is a simple function that can be computed quickly. Second, it avoids the problem of vanishing gradients for large input values, making deep neural networks easier to train. Third, it can help to sparsify network activation by setting some activations to zero, which can help to prevent overfitting.

```python
# def relu(x):
#     if x >= 0:
#         return x
#     else:
#         return 0
# y = list(map(relu, x))

y = list(map(lambda a: a if a>=0 else 0, x))

plot_graph(y,"ReLU(X)")
```

5.PRELU

The parametric rectified linear unit (PReLU) activation function is a ReLU variant that addresses the issue of "dying ReLU" by allowing for non-zero output for negative input values.

The PReLU function is superior to other activation functions in several ways. First, by allowing for non-zero output for negative input values, it can help to prevent the problem of dying ReLU. Second, it can help to improve network accuracy, particularly for datasets with a high level of noise or variability. Third, it is a simple function with a low computational cost.

```python
alpha = 0.5

y = list(map(lambda a: a if a>=0 else a*0.5, x))
plot_graph(y,"ReLU(X)")
```

6.ERELU

The Exponential Rectified Linear Unit (eReLU) activation function is a variant of the widely used ReLU activation function in artificial neural networks. The eReLU function is intended to address the "dying ReLU" issue, which occurs when ReLU neurons with negative inputs become "stuck" at 0 and cease to learn.

```python
import numpy as np

def eReLU(x, alpha=0.05):
    """ Exponential Rectified Linear Unit activation function """
    return np.where(x > 0, x, alpha*(np.exp(x) - 1))

x = np.linspace(-5, 5, 40)
y = list(map(lambda a: eReLU(a), x))

plot_graph(y, "eReLU(X)")
```
